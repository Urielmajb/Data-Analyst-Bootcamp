# If - Elif - Else Statements



```python
if 25 > 10:
    print('It worked')
```

```python
if 25 < 10:
    print('It worked')
else:
    print('It didnt work lol')
```

```python
if 25 > 10:
    print('It worked')
else:
    print('It didnt work lol')
```

```python
if (25 < 10) or (1 < 3):
    print('It worked')
elif 25 < 20:
    print('Elif worked')
elif 25 < 21:
    print('Elif 2 worked')
elif 25 < 40:
    print('Elif 3 worked')
elif 25 < 50:
    print('Elif 3 worked')
else:
    print('It didnt work lol')
```

```python
print('It worked!') if 10 > 30 else print('It didnt work lol') # Tambien se puede escribir en una sola linea
```

```python
if (25 < 10) or (1 < 3):
    print('It worked')
    if 10 > 5:
        print('THIS NESTED IF STATEMENT WORKED')
elif 25 < 20:
    print('Elif worked')
elif 25 < 21:
    print('Elif 2 worked')
elif 25 < 40:
    print('Elif 3 worked')
elif 25 < 50:
    print('Elif 3 worked')
else:
    print('It didnt work lol')
```
