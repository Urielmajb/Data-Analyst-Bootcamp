# For Loops



```python
integers = [1,2,3,4,5]

for numbers in integers:
    print(numbers)
```

```python
for numbers in integers:
    print('yeap')
```

```python
integers = [1,2,3,4,5]

for jelly in integers:
    print(jelly + jelly)
```

```python
ice_cream_dict= {'name': 'Alex Freberg', 'Weekly intake': 5, 'Favorite ice creams': ['MCC', 'Chocolate']}

for cream in ice_cream_dict.values():
    print(cream)
```

```python
ice_cream_dict= {'name': 'Alex Freberg', 'Weekly intake': 5, 'Favorite ice creams': ['MCC', 'Chocolate']}

for cream in ice_cream_dict.keys():
    print(cream)
```

```python
ice_cream_dict= {'name': 'Alex Freberg', 'Weekly intake': 5, 'Favorite ice creams': ['MCC', 'Chocolate']}

for key,value in ice_cream_dict.items():
    print(key,'->', value)
```

# Bucles Anidados


```python
flavor = ['Vanilla', 'Chocolate', 'Cookie Dough']
toppings= ['Hot Fudge', 'Oreos', 'Marshmallows']

for one in flavor:
    for two in toppings:
        print(one, 'topped with', two)
```
