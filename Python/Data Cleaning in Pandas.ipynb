# Data cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\uriel\Escritorio\Bootcamp Data Analyst\Python\Customer Call List.xlsx")
df
```

```python
# Escenario: Un leader de CS, quiere hacer llamadas con la siguiente lista, pero hay muchos datos duplicados y/o incoherentes, por lo tanto quiere que se le haga
# limpieza y entregarsela para que hagan las llamadas lo mas rapido posible.
```

```python
# Eliminamos los datos duplicados

df = df.drop_duplicates() 
df
```

```python
# Borramos columnas que no tengan sentido 
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df
```

```python
# Eliminar caracteres que tengan demas

df["Last_Name"] = df["Last_Name"].str.lstrip("...") # Se puede hacer de manera individual pero seria mas tequio 
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")

#De manera general, para quitar caracteres 
#df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

```python
df
```

```python
# Formateamos los numeros de celulares
#df["Phone_Number"]= df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True )
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#str(df["Phone_Number"])

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df["Phone_Number"]= df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]= df["Phone_Number"].str.replace('Na--','')
```

```python
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True) #Separa los datos. 
df
```

```python
#df["Paying Customer"]= df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"]= df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df["Do_Not_Contact"]= df["Paying Customer"].str.replace('Yes', 'Y')
df["Do_Not_Contact"]= df["Paying Customer"].str.replace('No', 'N')
df
```

```python
#df = df.replace('N/a', '')
#df = df.replace('NaN', '')
df = df.fillna('')
df
```

```python
# Tambein nos pidieron filtrar los numeros a los que se pueden llamar

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

```

```python
df = df.reset_index(drop=True)
df
```
